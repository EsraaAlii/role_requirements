```python
MLFLOW_TRACKING_URI = "../models/mlruns"
MLFLOW_RUN_ID = "d289e1d736954b17addda6c390b2f540"
EXP_ID = "117654367148467074"
CLUSTERS_YAML_PATH = "../data/processed/features_skills_clusters_description.yaml"
```

```python
import pandas as pd
import sys

sys.path.append('../')
from scripts.JobPrediction import *
```

### Simulate additional skills

#### 1-Calc base probs

```python
current_skills = ['C', 'C++']
#current_skills = ['R', 'SQL', 'RStudio']
#current_skills = ['Python', 'Pandas']
```

```python
model = JobPrediction(MLFLOW_TRACKING_URI ,MLFLOW_RUN_ID ,EXP_ID ,CLUSTERS_YAML_PATH)
```

```python
#calc bas probs 
base_pred = model.predict_jobs_probs(current_skills)
base_pred.sort_values(ascending=False)
```

#### 2-Calc probs for new skills

```python
#get possible new skills
all_skills = pd.Series(model.get_all_skills())
new_skills = all_skills[~all_skills.isin(current_skills)].copy()
```

```python
new_skills
```

```python
simulated_res = []
for skill in new_skills:
    additional_skill_prob = model.predict_jobs_probs([skill] + current_skills)
    additional_skill_uplift = (additional_skill_prob - base_pred)/base_pred
    additional_skill_uplift.name = skill
    simulated_res.append(additional_skill_uplift)
```

```python
simulated_res = pd.DataFrame(simulated_res)
```

```python
simulated_res
```

### Recommend

```python
target_job = 'Data scientist or machine learning specialist'
```

```python
target_res = simulated_res[target_job].sort_values(ascending=False)
target_res
```

```python
threshold = 1
recommendations = target_res[target_res>threshold].index.tolist()
```
