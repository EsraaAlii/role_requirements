```python
#packages
import pandas as pd
import numpy as np
import logging
import pickle
```

```python
DATA_PATH = r"D:\role_requirements\data\raw\survey_results_public.csv"
EXPORT_PATH = r"D:\role_requirements\data\processed\preprocessed_df.pkl"

REPLACE_DICT ={
    "YearsCode" : {'More than 50 years' : 51 , 'Less than 1 year' : 0},
    "YearsCodePro" : {'More than 50 years' : 51 , 'Less than 1 year' : 0},
}
```

```python
def split_answers(df):
    """
    split multiple answers to list
    """
    cols_with_multiple_answers = ['NEWCollabToolsWantToWorkWith','NEWCollabToolsHaveWorkedWith','ToolsTechWantToWorkWith','ToolsTechHaveWorkedWith','MiscTechWantToWorkWith','MiscTechHaveWorkedWith','WebframeWantToWorkWith','WebframeHaveWorkedWith','PlatformWantToWorkWith','PlatformHaveWorkedWith','DatabaseWantToWorkWith','DatabaseHaveWorkedWith','LanguageWantToWorkWith','LanguageHaveWorkedWith','DevType']

    for col in cols_with_multiple_answers:
        if ";":
            df[col] = df[col].str.split(';')
        # elif ",":
        #     df[col] = df[col].str.split(',')
        musk_null = df[col].isnull()
        df[col].loc[musk_null] = df[col].loc[musk_null].apply(lambda x : [])
    return df
        
```

## Preprocess Data

```python
raw_df = pd.read_csv(DATA_PATH)
df = raw_df.copy()
```

```python
#replace values
for col , replacement in REPLACE_DICT.items():
    df[col] = df[col].replace(replacement).astype("float")
```

```python
#split multiple answers 
df = split_answers(df)
```
